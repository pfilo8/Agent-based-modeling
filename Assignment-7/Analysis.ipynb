```python
import glob
from itertools import product

import numpy as np
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

```python
graphs = ['complete']
ps = list(map(str, np.arange(0, 1.05, 0.05)))
qs = ['4']
fs = list(map(str, [0.2, 0.3, 0.4, 0.5]))
```

```python
results = []

for graph, p, q, f in product(graphs, ps, qs, fs):  
    name = f'{graph}-{p}-{q}'
    
    paths = glob.glob(f'results/{graph}/q-{q}/p-{p}/f-{f}/*.csv')

    res = []
    for path in paths:
        print(path)
        dataframe = pd.read_csv(path)
        results.append((p, q, f, dataframe.mean(axis = 1).values[0]))
```

```python
df = pd.DataFrame(results, columns = ['p', 'q', 'f', 'Concetration'])
```

```python
fig = px.line(
    data_frame = df.groupby(['p', 'f'])['Concetration'].mean().reset_index(),
    x = 'p',
    y = 'Concetration',
    color = 'f'
)

fig.show()
```

```python
data = df.groupby(['p', 'f'])['Concetration'].mean().reset_index()

fig = go.Figure()

for f in df['f'].unique():
    dataframe = data[data['f'] == f]
    
    fig.add_trace(
        go.Scatter(
            x = dataframe['p'],
            y = dataframe['Concetration'],
            name = f'f = {f}',
            mode = 'lines+markers'
        )
    )
    
fig.update_layout(
    title = 'Concetration in q-voter models.',
    xaxis_title="p",
    yaxis_title="Concetration"
)
    
fig.write_html('results/Concetration.html')
```
